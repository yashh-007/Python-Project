```python
class CustomList:
    def __init__(self):
        self.items = []

    def append(self, item):
        """Add an item to the end of the list."""
        self.items.append(item)

    def extend(self, iterable):
        """Extend the list by appending elements from the iterable."""
        self.items.extend(iterable)

    def insert(self, index, item):
        """Insert an item at a given position."""
        self.items.insert(index, item)

    def remove(self, item):
        """Remove the first occurrence of an item from the list."""
        self.items.remove(item)

    def pop(self, index=-1):
        """Remove and return an item at the given position in the list."""
        return self.items.pop(index)

    def clear(self):
        """Remove all items from the list."""
        self.items.clear()

    def index(self, item):
        """Return the index of the first occurrence of an item."""
        return self.items.index(item)

    def count(self, item):
        """Return the number of occurrences of an item in the list."""
        return self.items.count(item)

    def sort(self, reverse=False):
        """Sort the items of the list in place."""
        self.items.sort(reverse=reverse)

    def reverse(self):
        """Reverse the elements of the list in place."""
        self.items.reverse()

    def copy(self):
        """Return a shallow copy of the list."""
        return self.items.copy()

    def __str__(self):
        """Return a string representation of the list."""
        return str(self.items)

# Example usage
if __name__ == "__main__":
    custom_list = CustomList()

    custom_list.append(1)
    custom_list.append(2)
    custom_list.append(3)
    print("After appending:", custom_list)

    custom_list.extend([4, 5, 6])
    print("After extending:", custom_list)

    custom_list.insert(1, 10)
    print("After inserting:", custom_list)

    custom_list.remove(10)
    print("After removing 10:", custom_list)

    popped_item = custom_list.pop()
    print("Popped item:", popped_item)
    print("After popping:", custom_list)

    index_of_3 = custom_list.index(3)
    print("Index of 3:", index_of_3)

    count_of_2 = custom_list.count(2)
    print("Count of 2:", count_of_2)

    custom_list.sort()
    print("After sorting:", custom_list)

    custom_list.reverse()
    print("After reversing:", custom_list)

    copied_list = custom_list.copy()
    print("Copied list:", copied_list)

    custom_list.clear()
    print("After clearing:", custom_list)
```
